# Steps to solve an ML problem

## Step 1-Problem Statement-

Iris data set contains the sepal length and width and petal length and width of the flowers of the Iris species based on which the data is classified into Iris-setosa, Iris-virginica and Iris-versicolor. This is a classification problem where for a given sepal width, length and petal width, length we have to classify the flower as Iris-setosa, Iris-virginica or Iris-versicolor

## Step 2- Under standing the Frame

#### Features X:

sepal length (slen)

sepal width (swid)

petal length (plen)

petal width (pwid)


#### Output(Target) y:

class

## Step 3- Acquiring the dataset

```python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as  np
import pandas as pd

iris = pd.read_csv("D:\Python data structures\Iris_missing.csv")
iris
```

## Step 4- Refine

```python
# check for missing values
iris.isnull().sum()
```

```python
# filling missing values using statistical methods

iris.slen=iris.slen.fillna(np.mean(iris.slen))
iris.plen=iris.plen.fillna(np.mean(iris.plen))
iris.swid=iris.swid.fillna(np.mean(iris.swid))
iris.pwid=iris.pwid.fillna(np.mean(iris.pwid))

iris.isnull().sum()
```

## Step 5- Explore the Data set

## Step  6- Transform

```python
iris.dtypes
```

We need to convert 'class' column from string to numeric value. We will use Sci-kit Learn's Label Encoder for the same.

```python
iris_encoded=iris.copy()
```

```python
# Label Encoding
from sklearn.preprocessing import LabelEncoder
le_class=LabelEncoder()
le_class.fit(iris['class'])
iris_encoded['class']=le_class.transform(iris['class'])
iris_encoded.head()
```

## Step 7- Build the model

```python
# splitting the data set into attributes and outputs
X=iris_encoded.iloc[:, 0:-1]
y=iris_encoded['class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
# building the model
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score
```

```python
iris_tree= DecisionTreeClassifier(max_depth=5)
iris_tree.fit(X_train,y_train)
```

## Step 8- Validate the data set

The data set is split into training and validation data sets and k-fold cross validation is used where k, which is the number of splits is 5

```python
def val_kfold(iris_df,k):
    # k subsets of the training data set
    X_kfold=StratifiedKFold(n_splits=k)
    
    k_train_acc=[]
    k_val_acc=[]
    
    for t_ind,val_ind in X_kfold.split(X_train,y_train):
        # training data set
        iris_df=iris_df.fit(X_train.iloc[t_ind],y_train.iloc[t_ind])
        
        #predict for test(validation) data  set
        iris_val_predict=iris_df.predict(X_train.iloc[val_ind])
        
        # calculate accuracy
        iris_train_predict=iris_df.predict(X_train.iloc[t_ind])
        
        train_acc=accuracy_score(y_train.iloc[t_ind],iris_train_predict)
        val_acc=accuracy_score(y_train.iloc[val_ind],iris_val_predict)
        
        print(train_acc, val_acc)
        
        k_train_acc.append(train_acc)  
        k_val_acc.append(val_acc)
    print('Mean of the k-fold accuracies-')
    
    k_train_mean=np.mean(k_train_acc)
    k_val_mean=np.mean(k_val_acc)
    return k_train_mean,k_val_mean
```

```python
val_kfold(iris_tree,5)
```

## Step 9- Tuning Hyper parameters
In our data set, we are getting a good training and validation accuracy. In case we do not get a good accuracy, we may change some hyper parameters such as the value of max_depth, or the value of k. Let's try with a max depth of 10

```python
iris_tree= DecisionTreeClassifier(max_depth=10)
iris_tree.fit(X_train,y_train)
val_kfold(iris_tree,5)
```

We are getting a perfect training set accuracy but a comparatively lower validation accuracy with max_depth of 10. So we can plot a graph to find out which max_depth will give the best accuracy and set the hyper parameter max_depth to the respective value

```python
train_acc_means=[]
val_acc_means=[]
depth=range(1,15)
for i in range (1,15):
    print(f'max depth={i}')
    iris_tree= DecisionTreeClassifier(max_depth=i)
    iris_tree.fit(X_train,y_train)
    each_train,each_val=val_kfold(iris_tree,5)
    train_acc_means.append(each_train)
    val_acc_means.append(each_val)
hyper_plot_df=pd.DataFrame({"depth":depth,
                       "train_acc_means":train_acc_means,
                       "val_acc_means":val_acc_means})
plt.plot(depth,train_acc_means, label='Training Accuracy')
plt.plot(depth,val_acc_means, label='Testing Accuracy')
plt.xlabel('depth')
plt.ylabel('accuracy')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

```

## Step 10- Testing

The model has to be tested against an unseen input. We will use X_test which we has originally separated in step 7

```python
X_test=X_test.drop(['Predicted Class'],axis=1)
X_test=X_test.drop(['Actual Class'],axis=1)
iris_tree_test= DecisionTreeClassifier(max_depth=4)
iris_tree_test.fit(X_train,y_train)
iris_test_predict=iris_tree_test.predict(X_test)
X_Test_op=X_test
X_Test_op['Actual Class']=y_test
X_Test_op['Predicted Class']=iris_test_predict

```
